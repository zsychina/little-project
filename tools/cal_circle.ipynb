```python
# calculate a circle to display in OLED
import numpy as np
np.set_printoptions(threshold = 1e6)
GRAM=np.zeros((128,64),dtype=np.int32)
cnt=0

for x in range(0,128):
    for y in range(0,64):
        if abs( (x-64)**2+(y-64)**2-64**2 ) <= 30:
            GRAM[x][y]=1
            cnt+=1
            # print(x,y)
            print('  OLED_DrawPoint({},{},1);'.format(x,y))

# print('all %d counts'%cnt)
```
